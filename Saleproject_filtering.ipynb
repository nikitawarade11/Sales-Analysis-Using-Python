# Data Filtering

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"D:\UV DA\UV python\Sales_Project_Data.xlsx")
```

```python
df
```

```python
df[df['Sales'] > 1000]
```

```python
df[df['Profit'] < 200]
```

```python
df[df['Region'] == 'East']
```

```python
df[(df['Sales'] > 1000) & (df['Profit'] > 200)]
```

```python
df[(df['Region'] == 'East') & (df['Category'] == 'Electronics')]
```

```python
df[df['Region'].isin(['East', 'West'])]
```

```python
df[df['Category'].isin(['Furniture', 'Clothing'])]
```

```python
df[df['Product'].str.contains('Laptop')]
```

```python
df[df['Region'].str.startswith('N')]
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df[df['Date'].dt.month == 1]
```

```python
df.nlargest(5, 'Sales')
```

```python
df.nsmallest(5, 'Profit')
```

```python
df['Profit_Percentage'] = (df['Profit'] / df['Sales']) * 100
```

```python
df[df['Profit_Percentage'] > 20]
```

```python
df.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'mean'
})
```

```python
df.groupby('Product')['Sales'].sum().sort_values(ascending=False).head(3)
```

```python
pd.pivot_table(df,
               values='Sales',
               index='Region',
               columns='Category',
               aggfunc='sum',
               fill_value=0)
```

```python
df.loc[df.groupby('Region')['Sales'].idxmax()]
```
