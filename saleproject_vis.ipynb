# Matplotlib 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
ab = pd.read_excel(r"D:\UV DA\UV python\Sales_Project_Data.xlsx")
```

```python
ab
```

```python

ab['Date'] = pd.to_datetime(ab['Date'])

plt.figure()
plt.plot(ab['Date'], ab['Sales'], color='red', marker='^')

plt.xticks(rotation=45)
plt.title("Sales Over Time")
plt.xlabel("Date", fontsize=12, fontweight='bold')
plt.ylabel("Sales", fontsize=12, fontweight='bold')
plt.grid()

plt.show()
```

```python
col = ['pink','purple','green','yellow']
plt.bar(ab['Region'],ab['Sales'],color= col)
plt.title("Sale by Regiom")
plt.xlabel('Region',fontsize = 12,fontweight= 'bold')
plt.ylabel("Sales",fontsize = 12,fontweight= 'bold')

plt.show()
```

```python
plt.figure()
plt.scatter(ab["Quantity"], ab["Sales"])
plt.title("Sales by Quantiti" ,fontsize = 12 , fontweight = 'bold')
plt.xlabel("Quantity",fontsize = 12, fontweight = 'bold')
plt.ylabel("Sales",fontsize = 12, fontweight = 'bold')
plt.show()
```

```python
plt.figure(figsize=(6,4))

scatter = plt.scatter(
    ab['Quantity'], 
    ab['Sales'],
    s=ab['Sales'],        # Bubble size (scaled)
    c=ab['Sales'],            # Color based on Sales
    cmap='cool',          #color theme('viridis''plasma''cool''hot''Set1')
    alpha=0.6,              #colour transperancy(0 → fully transparent,1 → fully solid,0.6 → semi transparent)
    edgecolors='black'
)

plt.colorbar(scatter, label="Sales")
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.title("Bubble Chart: Quantity vs Sales")

plt.show()
```

```python
plt.hist(ab['Sales'],bins=9,color='y')
plt.show()
```

```python
pivot = ab.pivot_table(index = 'Region',columns='Product' ,values='Sales' , aggfunc= 'sum')

pivot.plot(kind = 'bar' , stacked = False,color = ['r','y','g'])

plt.show()
```

```python
pivot = ab.pivot_table(index = 'Region',columns='Product' ,values='Sales' , aggfunc= 'sum')

pivot.plot(kind = 'bar' , stacked = True)

plt.show()
```

```python
grouped_ab= ab.groupby('Region')['Sales'].sum()
plt.pie(grouped_ab, labels= grouped_ab.index, autopct='%1.1f%%') #1f%% show decimal place
plt.show()
```

```python
sb.boxplot(x='Region', y='Sales', data=ab,color = 'y')

plt.title("Sales by Region")
plt.show()
```

```python
sb.countplot(data=ab, x='Product',palette=['r','y','g'])
plt.title("Product Count",fontsize = 10,fontweight='bold',color = 'green')
plt.show()
```

```python
sb.scatterplot(x='Product', y='Sales' , data = ab , hue = 'Category')
plt.title("Sales vs Quantity")

plt.show()
```

```python
plt.figure(figsize=(8,6))
sb.lineplot(data=ab, x='Date', y='Sales',color = 'g')
plt.xticks(rotation=45)
plt.title("Sales Over Time")
plt.show()
```

```python
sb.barplot(data=ab, x='Region', y='Sales', hue='Product',errorbar = None)
plt.title("Region vs Product Sales")
plt.show()
```

```python
sb.histplot(data=ab, x='Sales', element='poly',color='y')
plt.show()
```

```python
sb.histplot(data=ab, x='Sales', hue='Region', multiple='layer')
plt.show()
```

```python
sb.kdeplot(data=ab, x='Sales', fill=True,color='green')
plt.title("Sales Density")
plt.show()
```

```python
sb.kdeplot(data=ab, x='Sales', hue='Region', fill=True)
plt.show()
```
